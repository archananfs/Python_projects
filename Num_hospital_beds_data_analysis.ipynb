```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('world-development-indicators\Indicators.csv')
#print(data.shape)
indicatorname = data['IndicatorName'].unique().tolist()
#len(indicatorname)
#indicatorname[0:10]
```

```python
countryfilter = ['IND', 'USA', 'ARB', 'BRA', 'CHN', 'DEU']
#countryfilter = 'IND'
indicatorfilter = 'Hospital beds \(per 1,000 '
mask1 = data['IndicatorName'].str.contains(indicatorfilter)
mask2 = data['CountryCode'].isin(countryfilter)
stage = data[mask1 & mask2]
```

```python
Y=[]
X=[]
y_for_label =pd.DataFrame([])
for i in range(len(countryfilter)):
    y_data1 = stage[stage['CountryCode'] == countryfilter[i]]
    y_data_val = y_data1['Value'].values
    Y.append(y_data_val)
    x_data_val = y_data1['Year'].values
    X.append(x_data_val)

df = pd.DataFrame({'CountryName': countryfilter})
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
for x_arr,y_arr, label in zip(X,Y,countryfilter):
    ax.plot(x_arr,y_arr, label=label)
plt.ylabel(stage['IndicatorName'].iloc[0])
plt.xlabel('Years')

legend = plt.legend(loc = 'upper right', 
                    shadow=False,
                    prop={'weight':'roman','size':'xx-large'})
# Rectangle around the legend
frame = legend.get_frame()
frame.set_facecolor('.95')
plt.show()
```
